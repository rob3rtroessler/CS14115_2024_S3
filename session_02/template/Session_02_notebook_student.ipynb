```python
# find date 
import re 

article_years = [] 

for preview in article_previews: 
    date_string_unprocessed = str(preview.find('a').get('href'))
    if 'article' in date_string_unprocessed: 
        # print(date_string_unprocessed)
        pattern = r'/article/(\d{4})/(\d{1,2})/(\d{1,2})/'
        year, month, day = re.search(pattern, date_string_unprocessed).groups()
        # only to extract the year 
        article_years.append(year)
article_years
```

## Task #2 - generate a file with information for the marketing manager

OK, now that you know how to retrieve this information, write a program that extracts this information and writes it into the csv file "article_previews.csv". If you can't quite remember how you create and write into files, take a look at yesterday's assignment.

```python
import pandas as pd 
import re 

data = [] 

for preview in article_previews: 
    title = preview.find("h1").text
    tag = preview.find("div", class_ = "subsection css-mtdwtq").text
    date_string_unprocessed = str(preview.find('a').get('href'))
    if 'article' in date_string_unprocessed: 
        # print(date_string_unprocessed)
        pattern = r'/article/(\d{4})/(\d{1,2})/(\d{1,2})/'
        year, month, day = re.search(pattern, date_string_unprocessed).groups()
    data.append([title, tag, year])

df = pd.DataFrame(data, columns=['Title', 'Tag', 'Year']) 

df.to_csv('article_previews2024.csv') 
df 
```
